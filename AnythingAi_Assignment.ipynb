```python
position_groups = merged_data.groupby(['Account', 'Coin', 'Start Position', 'Direction','Date']).agg({
    'Size Tokens': 'sum',
    'Size USD': 'sum',
    'Closed PnL': 'sum'
}).reset_index()
```

```python

merged_data['Notional_USD'] = merged_data['Size Tokens'].abs() * merged_data['Execution Price']
# Sort and compute price change proxy per position (simplified; refine with timestamps)
merged_data_sorted = merged_data.sort_values(['Account', 'Coin', 'Timestamp IST', 'Side'])
merged_data['Price_Change_Pct'] = merged_data.groupby(['Account', 'Coin'])['Execution Price'].pct_change().abs()
 #(current_price - previous_price) / previous_price


```

```python
merged_data.head(5)
```

```python
position_summary = merged_data[merged_data['Closed PnL'].notna()].groupby(['Account', 'Coin', 'Trade ID']).agg({
    'Closed PnL': 'sum',
    'Notional_USD': 'mean',
    'Price_Change_Pct': 'mean'
}).reset_index()
```

```python
position_summary['Proxy_Leverage'] = np.abs(position_summary['Closed PnL']) / (
    position_summary['Notional_USD'] * position_summary['Price_Change_Pct']
).replace(0, np.nan)
```

```python
# Extract unique Trade ID and their corresponding classification from merged_data
trade_id_classification = merged_data[['Trade ID', 'classification']].drop_duplicates(subset=['Trade ID'])

# Merge classification into position_summary
position_summary_with_classification = pd.merge(
    position_summary,
    trade_id_classification,
    on='Trade ID',
    how='left'
)

# Now, group by classification on the new DataFrame
behavior_summary = position_summary_with_classification.groupby('classification').agg({
    'Closed PnL': 'sum',
    'Account': 'count', # This will count the number of Trade IDs (positions) per classification
    'Proxy_Leverage' :'mean'
})
```

```python
behavior_summary
```

```python
average_leverage_greed = behavior_summary['Proxy_Leverage']['Greed']
average_leverage_fear = behavior_summary['Proxy_Leverage']['Fear']
```

```python
if average_leverage_greed > average_leverage_fear:
    print("Insight: Traders show increased risk-taking behavior during Greed phases.")
```
